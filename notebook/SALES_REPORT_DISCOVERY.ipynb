```python
def close_connection(curr, ctx):
  if curr:
    curr.close()
  if ctx:
    ctx.close()
```

```python
from google.colab import userdata
import snowflake.connector
```

```python
SNOWFLAKE_ACCOUNT_NAME = userdata.get('SNOWFLAKE_ACCOUNT_NAME')
SNOWFLAKE_DATABASE="demo"
SNOWFLAKE_PASSWORD=userdata.get("SNOWFLAKE_PASSWORD")
SNOWFLAKE_USER_NAME=userdata.get("SNOWFLAKE_USER_NAME")
SNOWFLAKE_WAREHOUSE="PROD_DEFAULT_WH"
SNOWFLAKE_TEMP_TABLE = "demo.hiring_temp.sales_ms"
SNOWFLAKE_PROD_TABLE = "demo.hiring_prod.sales_ms"
SNOWFLAKE_SCHEMA_TEMP = "hiring_temp"
```

```python

try:
    ctx = snowflake.connector.connect(
        user=SNOWFLAKE_USER_NAME,
        password=SNOWFLAKE_PASSWORD,
        account=SNOWFLAKE_ACCOUNT_NAME,
        warehouse=SNOWFLAKE_WAREHOUSE,
        database=SNOWFLAKE_DATABASE)
    curr = ctx.cursor()
    q = f"""
      SELECT
        *
      FROM
        {SNOWFLAKE_TEMP_TABLE}"""
    print(q)
    curr.execute(q)
    data = curr.fetchall()
    columns = [desc[0] for desc in curr.description]
    df_temp = pd.DataFrame(data, columns=columns)
except Exception as e:
    print(f"Error connecting to Snowflake or executing query: {e}")
    close_connection(curr, ctx)
```

# Transforming query temp to production

```python
PROD_QUERY = f"""
SELECT
  CURRENT_TIMESTAMP() AS LOAD_TIMESTAMP,
  Provider AS PROVIDER  ,
  "Provider Country" AS PROVIDER_COUNTRY,
  SKU AS SKU  ,
  Developer AS DEVELOPER  ,
  Title AS TITLE  ,
  Version AS VERSION  ,
  "Product Type Identifier" AS PRODUCT_TYPE_IDENTIFIER,
  Units AS UNITS  ,
  "Developer Proceeds" AS DEVELOPER_PROCEEDS,
  TO_DATE("Begin Date") AS BEGIN_DATE,
  TO_DATE("End Date") AS END_DATE,
  "Customer Currency" AS CUSTOMER_CURRENCY,
  "Country Code" AS COUNTRY_CODE,
  "Currency of Proceeds" AS CURRENCY_OF_PROCEEDS,
  "Apple Identifier" AS APPLE_IDENTIFIER,
  "Customer Price" AS CUSTOMER_PRICE,
  "Promo Code" AS PROMO_CODE,
  "Parent Identifier" AS PARENT_IDENTIFIER,
  Subscription AS SUBSCRIPTION  ,
  Period AS PERIOD  ,
  Category AS CATEGORY  ,
  CMB AS CMB  ,
  Device AS DEVICE  ,
  "Supported Platforms" AS SUPPORTED_PLATFORMS,
  "Proceeds Reason" AS PROCEEDS_REASON,
  "Preserved Pricing" AS PRESERVED_PRICING,
  Client AS CLIENT  ,
  "Order Type" AS ORDER_TYPE
FROM
{SNOWFLAKE_TEMP_TABLE}
"""
```

```python
try:
    curr = ctx.cursor()
    curr.execute(PROD_QUERY)
    data = curr.fetchall()
    columns = [desc[0] for desc in curr.description]
    df_t = pd.DataFrame(data, columns=columns)
    curr.close()
except Exception as e:
    print(f"Error connecting to Snowflake or executing query: {e}")
    close_connection(curr, ctx)
```
